```python
import numpy as np
import pandas as pd
```

```python
# Importing the dataset
# The data set was first exported to excel as csv file
world = pd.read_csv('C:\\Users\\MY LAPTOP\\Desktop\\SQI\\Data Science\\Numpy\\Numpy Project Dataset\\world-d.csv')
world.head()
```

```python
# Creating np array for name, region and population
country_name = np.array(world['Name'])
country_region = np.array(world['Region'])
pop = np.array(world['Population'])
local_name = np.array(world['LocalName'])
gnp = np.array(world['GNP'])
```

# Country name and region with highest population

```python
# Country name with highest population
country_name[pop.argmax()]
```

```python
# Region with highest population
country_region[pop.argmax()]
```

# Get the localname with the highest and lowest GNP

```python
# np array of localname and GNP
print("Localname:\n", local_name, '\n')
print("GNP:\n", gnp)
```

```python
# Getting highest and lowest GNP
print("Highest GNP: ", gnp.max())
print("Lowest GNP: ", gnp.min())
```

```python
# Getting index of highest and lowest GNP
print("Index of Highest GNP: ", gnp.argmax())
print("Index of Lowest GNP: ", gnp.argmin())
```

```python
# Localname with highest GNP
local_name[gnp.argmax()]
```

```python
# Localname with lowest GNP
local_name[gnp.argmin()]
```

```python
# Confirming highest GNP by sorting
world.sort_values(by='GNP',ascending=False)
```

# Extract all the null and not null in both indepYear and GNPOld

```python
# Creating np array of IndepYear
indep = world['IndepYear'].values
indep
```

```python
# Extracting all the null in IndepYear
indep_nan = pd.isna(indep)
indep[indep_nan]
```

```python
# Extracting all values in IndepYear
indep_values = pd.notna(indep)
indep[indep_values]
```

```python
# Creating np array of GNPOld
gnpold = world['GNPOld'].values
gnpold
```

```python
# Extracting all the null in gnpold
gnpold_nan = pd.isna(gnpold)
gnpold[gnpold_nan]
```

```python
# Extracting all the values in gnpold
gnpold_values = pd.notna(gnpold)
gnpold[gnpold_values]
```
